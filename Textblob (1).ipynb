```python
import pandas as pd
from textblob import TextBlob
#Create a function to get the polarity
def getPolarity(text):
    return TextBlob(text).sentiment.polarity


def getAnalysis1(score):
    if score < 0:
        return 'NEGATIVE'
    elif score == 0:
        return 'NEUTRAL'
    else:
        return 'POSITIVE'

def getAnalysis(score):
    if score < 0:
        return 'NEGATIVE'    
    else:
        return 'POSITIVE'
```

```python
df=pd.read_csv("Adhoc_analysis - Utterance sentiment samples (1).csv")
# Adhoc_analysis - Utterance sentiment samples (1)
df.columns
df_new=df[['mtn', 'acct_no', 'ivrcallid', 'event_dt', 'chat_text', 'L1_cat',
    'L2_cat','L3_cat','rpt_date','processed_text','sentiment_score','sentiment']]
df_new.head()
```

```python
df_new['sentiment_score'] = df_new['processed_text'].apply(getPolarity)
df_new ['sentiment'] = df_new  ['sentiment_score'].apply(getAnalysis )
```
